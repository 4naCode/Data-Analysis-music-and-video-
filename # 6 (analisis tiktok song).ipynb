```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score


params = {'random_state': 0, 'n_jobs': 4, 'n_estimators': 5000, 'max_depth': 8}
# One-hot encode
drop = ['track_name', 'artist_name', 'album']
x, y = df.drop(drop,axis=1), df['artist_pop']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
# Fit RandomForest Classifier
clf = RandomForestClassifier(**params)
#clf = clf.fit(x, y)
# Fit the model
clf.fit(X_train, y_train)

# Make predictions
y_pred = clf.predict(X_test)

# Assuming you have y_test (true labels) and y_pred (predicted labels)
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

print(f'Accuracy: {accuracy}')

# Plot features importances
imp = pd.Series(data=clf.feature_importances_, index=x.columns).sort_values(ascending=False)
plt.figure(figsize=(10,12))
plt.title("Feature importance")
ax = sns.barplot(y=imp.index, x=imp.values, palette="Blues_d", orient='h')
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
drop = ['track_name', 'artist_name', 'album']
x, y = df.drop(drop,axis=1), df['artist_pop']

# Fit RandomForest Classifier
clr = RandomForestClassifier(**params)
clr = clr.fit(x, y)

# Plot features importances
imp = pd.Series(data=clr.feature_importances_, index=x.columns).sort_values(ascending=False)
plt.figure(figsize=(10,12))
plt.title("Feature importance")
ax = sns.barplot(y=imp.index, x=imp.values, palette="Blues_d", orient='h')
```
